2) 데이터프레임, 리스트, 딕셔너리, ndarray 상호 변환

① 리스트, ndarray, 딕셔너리 → 데이터프레임

- 2차원 이하만 가능



```python
import numpy as np

col_name1 = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 dataframe 생성
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 dataframe:\n', df_list1)

# numpy ndarray를 이용해 dataframe 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 dataframe:\n', df_array1)
```

```python
# 2행 3열 크기의 2차원 데이터 생성 위해 컬럼 명 3개 지정
col_name2 = ['col1', 'col2', 'col3']
list2 = [[1,2,3], [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 dataframe:\n', df_list2)

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 dataframe:\n', df_array2)
```

```python
# 딕셔너리를 dataframe으로 변환
dict = {'col1':[1,11], 'col2':[2, 22], 'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 dataframe:\n', df_dict)
```

② 데이터 프레임 → ndarray, 리스트, 딕셔너리

- 데이터 프레임 → ndarray : values 이용

- 데이터 프레임 → 리스트 : tolist() 이용

- 데이터 프레임 → 딕셔너리 : to_dict() 이용

```python
# dataframe을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)

# dataframe을 리스트로 변환
list3 = df_dict.values.tolist()
print('\ndf_dict.values.tolist() 타입:', type(list3))
print(list3)

# dataframe을 딕셔너리로 전환
dict3 = df_dict.to_dict('list')
print('\ndf_dict.todict() 타입:', type(dict3))
print(dict3)
```

9) apply lambda 식으로 데이터 가공

① lambda

- 파이썬에서 함수형 프로그래밍 지원

- 함수의 선언 & 함수 내 처리 를 식 1줄로 변환

- : 로 좌측 우측 구분. 좌측에는 입력 인자, 우측에는 반환 값(입력 인자의 계산식)

- 입력 인자 여러 개인 경우 : map() 함수 결합

- if else 지원 O : if 절의 경우 if 식보다 반환 값 먼저 기술, else절의 경우 else 식을 먼저 적은 후 반환 값 기술

- else if는 지원하지 x. else절을 ()로 내포해 () 내에서 다시 if else 적용해 사용해야 함

- else if 많이 나와야 하는 경우는 별도의 함수 생성하는 편이 나음

*apply : 데이터 프레임의 경우 map대신 apply() 함수 적용 가능

```python
# 입력값의 제곱 값을 반환하는 get_square(a) 함수
def get_square(a) :
  return a**2

print('3의 제곱은', get_square(3))
```

```python
lambda_square = lambda x : x**2
print('3의 제곱은', lambda_square(3))
```

```python
a=[1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
# 예제 : 나이에 따라 세분화된 분류 수행하는 함수 생성
def get_category(age):
  cat = ''
  if age <= 5 : cat = 'Baby'
  elif age <= 12 : cat = 'Child'
  elif age <= 18 : cat = 'Teenage'
  elif age <= 25 : cat = 'Student'
  elif age <= 35 : cat = 'Young Adult'
  elif age <= 60 : cat = 'Adult'
  else : cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
